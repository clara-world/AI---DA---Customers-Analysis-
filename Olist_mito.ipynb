```python
import pandas as pd

# File paths (assuming they are all in the same folder as your notebook)
data_path = "Jupyter/"  # or specify the full path if they're in another directory

# Read each CSV into a DataFrame
customers_df = pd.read_csv(f"{data_path}olist_customers_dataset.csv")
geolocation_df = pd.read_csv(f"{data_path}olist_geolocation_dataset.csv")
order_items_df = pd.read_csv(f"{data_path}olist_order_items_dataset.csv")
order_payments_df = pd.read_csv(f"{data_path}olist_order_payments_dataset.csv")
order_reviews_df = pd.read_csv(f"{data_path}olist_order_reviews_dataset.csv")
orders_df = pd.read_csv(f"{data_path}olist_orders_dataset.csv")
products_df = pd.read_csv(f"{data_path}olist_products_dataset.csv")
sellers_df = pd.read_csv(f"{data_path}olist_sellers_dataset.csv")
category_translation_df = pd.read_csv(f"{data_path}product_category_name_translation.csv")
```

```python
# Display column names for each dataframe
def display_dataframe_columns(dataframes):
    for df_name, df in dataframes.items():
        print(f"{df_name}:\n", df.columns.tolist(), "\n")

# Call the function to display columns
all_dataframes = {
    'customers_df': customers_df,
    'geolocation_df': geolocation_df,
    'orders_df': orders_df,
    'order_items_df': order_items_df,
    'order_payments_df': order_payments_df,
    'order_reviews_df': order_reviews_df,
    'products_df': products_df,
    'sellers_df': sellers_df,
    'category_translation_df': category_translation_df
}
display_dataframe_columns(all_dataframes)  
```

```python
# Clean city names in the 'customer_city' column
customers_df['customer_city_clean'] = (
    customers_df['customer_city']
    .str.strip()              # Remove leading/trailing spaces
    .str.lower()              # Convert to lowercase
    .str.normalize('NFKD')    # Normalize accented characters
    .str.encode('ascii', errors='ignore')  # Remove non-ASCII characters
    .str.decode('utf-8')      # Decode back to string
    .str.replace(r'\s+', ' ', regex=True)  # Replace multiple spaces with one
)

# Optional: Replace the original column if desired
# customers_df['customer_city'] = customers_df['customer_city_clean']
# del customers_df['customer_city_clean']

# Check the result
customers_df[['customer_city', 'customer_city_clean']].head()
```

```python
# Count number of records per city
city_counts = customers_df['customer_city_clean'].value_counts().reset_index()

# Rename columns for clarity
city_counts.columns = ['city', 'customer_count']

# Display the top 10 cities by number of customers
city_counts.head(100)
```

```python
# Step 1: Convert timestamp
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])

# Step 2: Merge with customers to get customer_unique_id
orders_merged = orders_df.merge(customers_df[['customer_id', 'customer_unique_id']], on='customer_id', how='left')

# Step 3: Count number of orders per unique customer
customer_order_counts = (
    orders_merged.groupby('customer_unique_id')['order_id']
    .nunique()
    .reset_index()
    .rename(columns={'order_id': 'order_count'})
)

# Step 4: Calculate retention rate
total_customers = customer_order_counts.shape[0]
repeat_customers = customer_order_counts[customer_order_counts['order_count'] > 1].shape[0]
retention_rate = repeat_customers / total_customers

print(f"Overall Retention Rate: {retention_rate:.2%}")
```

```python
# Add cohort month (first purchase)
first_order = (
    orders_merged.groupby('customer_unique_id')['order_purchase_timestamp']
    .min()
    .dt.to_period('M')
    .reset_index()
    .rename(columns={'order_purchase_timestamp': 'cohort_month'})
)

# Add to customer orders
orders_with_cohort = orders_merged.merge(first_order, on='customer_unique_id')

orders_with_cohort
```

```python
orders_with_cohort['order_count'] = orders_with_cohort.groupby('customer_unique_id')['order_id'].transform('count')

customers_with_multiple_orders = orders_with_cohort[orders_with_cohort['order_count'] > 1]

repeat_customer_count_by_cohort = (
    customers_with_multiple_orders.groupby('cohort_month')['customer_unique_id']
    .nunique()
    .reset_index()
    .rename(columns={'customer_unique_id': 'repeat_customer_count'})
)

repeat_customer_count_by_cohort
```

```python
customer_counts_by_cohort = first_order.groupby('cohort_month')['customer_unique_id'].nunique()
customer_counts_by_cohort_df = customer_counts_by_cohort.reset_index().rename(columns={0: 'cohort_count'})
customer_counts_by_cohort_df .rename(columns={'customer_unique_id': 'customer_unique_count'}, inplace=True)

customer_counts_by_cohort_df 
```

```python
# Perform inner join on 'cohort_month'
merged_cohorts = pd.merge(customer_counts_by_cohort_df, repeat_customer_count_by_cohort, on='cohort_month')

# Calculate retention rate
merged_cohorts['retention_rate'] = (
    merged_cohorts['repeat_customer_count'] / merged_cohorts['customer_unique_count']
)

merged_cohorts
```

```python
# Compute mean and standard deviation of retention_rate
mean_retention = merged_cohorts['customer_unique_count'].mean()
std_retention = merged_cohorts['customer_unique_count'].std()

# Define outlier threshold (remove below mean - 2*std)
threshold = mean_retention - std_retention

# Filter out rows with outlier retention rates
filtered_cohorts = merged_cohorts[merged_cohorts['customer_unique_count'] >= threshold]

# Drop the 'index' column if it exists
filtered_cohorts = filtered_cohorts.drop(columns=['index'], errors='ignore')

# Reset index for clean output
filtered_cohorts_df = filtered_cohorts.reset_index(drop=True)

# Show result
filtered_cohorts_df
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(filtered_cohorts_df['cohort_month'].astype(str), filtered_cohorts_df['retention_rate'], color='skyblue')

# Add labels and title
plt.xlabel('Cohort Month')
plt.ylabel('Retention Rate')
plt.title('Retention Rate by Cohort Month')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
