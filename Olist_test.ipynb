```python
import pandas as pd

# Assuming 'merged_df' contains the merged data
# and 'order_purchase_timestamp' is the order date column

# Convert 'order_purchase_timestamp' to datetime before calculating recency
merged_df = pd.merge(merged_df, order_item_df, on='order_id', how='left')
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])

# Calculate Recency
recency = (pd.Timestamp.now() - merged_df.groupby('customer_unique_id')['order_purchase_timestamp'].max()).dt.days

# Calculate Frequency
frequency = merged_df.groupby('customer_unique_id')['order_id'].nunique()

# Calculate Monetary Value
monetary_value = merged_df.groupby('customer_unique_id')['price'].sum()

# Create RFM DataFrame
rfm_df = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'MonetaryValue': monetary_value})
rfm_df = rfm_df.index.to_frame().join(rfm_df)  # Add customer_unique_id as a column

# Display the RFM DataFrame
display(rfm_df.head())

# Further analysis and segmentation can be performed based on RFM values
```
